```python
import pandas
```

```python
#inputs
x = 4
months = 0.5
interest = 1.5
BTC_ETH_1 = 0.069195
USD_ETH_1 = 250
```

```python
def calc(x,months,interest,BTC_ETH_1,USD_ETH_1,BTC_ETH_2,USD_ETH_2):
    multiplier = (1 + interest/100)**months
    int_payment = (x * USD_ETH_1) - (x * USD_ETH_1) * multiplier
    BTC_x_1 = x * BTC_ETH_1
    ETH_x_2 = BTC_x_1 / BTC_ETH_2
    ETH_p_2 = ETH_x_2 * USD_ETH_2
    ETH_p_1_2 = x * USD_ETH_2
    net_prof = ETH_p_2 - ETH_p_1_2 + int_payment
    return net_prof
```

```python
df=pandas.read_csv('/Users/edwardhuang/Documents/python/BTC/test.csv', header=None)
df
```

```python
rows = df.shape[0]
cols = df.shape[1]
```

```python
for n in range (1,rows):
    for y in range (1,cols):
        df.iloc[n,y] = calc(x,months,interest,BTC_ETH_1,USD_ETH_1,df.iloc[0,y],df.iloc[n,0])
```
